## Libraries

```python
# Required libraries
import pandas as pd
import os
import json
import pprint
```

## Data Fetching

```python
# Aggregated-Transaction
path = "pulse/data/aggregated/transaction/country/india/state/"

Agg_state_list = os.listdir(path)
Agg_state_list

clm={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list:
    p_i = path + i + "/"
    Agg_yr_list = os.listdir(p_i)
    for j in Agg_yr_list:
        p_j = p_i + j + "/"
        Agg_Q_list = os.listdir(p_j)
        for k in Agg_Q_list:
            p_k = p_j + k
            Data = open(p_k,'r')
            D = json.load(Data)
            for z in D['data']['transactionData']:
                Name = z['name']
                count = z['paymentInstruments'][0]['count']
                amount = z['paymentInstruments'][0]['amount']
                clm['Transaction_type'].append(Name)
                clm['Transaction_count'].append(count)
                clm['Transaction_amount'].append(amount)
                clm['State'].append(i)
                clm['Year'].append(j)
                clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Trans = pd.DataFrame(clm)
```

```python
Agg_Trans.to_csv("Agg_Trans.csv", encoding='utf-8', index=False)
```

```python
Agg_Trans
```

```python
# Aggregated-User
path = "pulse/data/aggregated/user/country/india/state/"

Agg_state_list = os.listdir(path)

clm={'State':[], 'Year':[],'Quater':[],'Registered_users':[], 'App_opens':[], 'Device_Brand':[], 'Brand_users':[], 'Device_share':[]}

for i in Agg_state_list:
    p_i = path + i + "/"
    Agg_yr_list = os.listdir(p_i)
    for j in Agg_yr_list:
        p_j = p_i + j + "/"
        Agg_Q_list = os.listdir(p_j)
        for k in Agg_Q_list:
            p_k = p_j + k
            Data = open(p_k,'r')
            D = json.load(Data)
            if D["data"]["usersByDevice"] is not None:
                for z in D["data"]["usersByDevice"]:
                    users = D["data"]["aggregated"]["registeredUsers"]
                    usage = D["data"]["aggregated"]["appOpens"]
                    Brand = z['brand']
                    Count = z['count']
                    percent = z['percentage']
                    clm['Registered_users'].append(users)
                    clm['App_opens'].append(usage)
                    clm['Device_Brand'].append(Brand)
                    clm['Brand_users'].append(Count)
                    clm['Device_share'].append(percent)
                    clm['State'].append(i)
                    clm['Year'].append(j)
                    clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Users = pd.DataFrame(clm)
```

```python
Agg_Users.to_csv("Agg_Users.csv", encoding='utf-8', index=False)
```

```python
# Top-Transaction
path = "pulse/data/top/transaction/country/india/state/"

Top_state_list = os.listdir(path)

clm={'State':[], 'District':[], "Pincode":[], 'Year':[], 'Quater':[], 'Trans_dist_count':[], 'Trans_dist_amount':[], 'Trans_pincode_count':[], 'Trans_pincode_amount':[]}

for i in Top_state_list:
    p_i = path + i + "/"
    Top_yr_list = os.listdir(p_i)
    for j in Top_yr_list:
        p_j = p_i + j + "/"
        Top_Q_list = os.listdir(p_j)
        for k in Top_Q_list:
            p_k = p_j + k
            Data = open(p_k,'r')
            D = json.load(Data)
            # for z in D['data']:
                # state = D['data']['states'][0]['entityName']
                # s_count = z['states'][0]['metric']['count']
                # s_amount = z['states'][0]['metric']['amount']
            district = D['data']["districts"][0]['entityName']
            d_count = D['data']["districts"][0]['metric']['count']
            d_amount = D['data']["districts"][0]['metric']['amount']
            pincode = D['data']["pincodes"][0]['entityName']
            p_count = D['data']["pincodes"][0]['metric']['count']
            p_amount = D['data']["pincodes"][0]['metric']['amount']
            clm['Trans_dist_count'].append(d_count)
            clm['Trans_dist_amount'].append(d_amount)
            clm['Trans_pincode_count'].append(p_count)
            clm['Trans_pincode_amount'].append(p_amount)
            clm['Pincode'].append(pincode)
            clm['District'].append(district)
            clm['State'].append(i)
            clm['Year'].append(j)
            clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Top_Trans = pd.DataFrame(clm)
```
